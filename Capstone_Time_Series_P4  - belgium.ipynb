# Checking for null values

```python
df[df.isnull().any(axis=1)]
```

```python
df["Neighbourhood"] = df["Neighbourhood"].fillna(df["Neighbourhood"].mode()[0])
df["Last Review"] = df["Last Review"].fillna(df["Last Review"].mode()[0])
df["Host Response Time"] = df["Host Response Time"].fillna(df["Host Response Time"].mode()[0])
df["Host Response Rate"] = df["Host Response Rate"].fillna(df["Host Response Rate"].mean())
df["Reviews per Month"] = df["Reviews per Month"].fillna(df["Reviews per Month"].mean())                                                                 
df["Security Deposit"] = df["Security Deposit"].fillna(df["Security Deposit"].mean())
df["Cleaning Fee"] = df["Cleaning Fee"].fillna(df["Cleaning Fee"].mean())
```

```python
Monthly = df['Price'].resample('m').mean()
```

```python
Monthly = pd.DataFrame(data = Monthly)
```

# Checking the Stationarity of the dataset

```python
from statsmodels.tsa.stattools import adfuller
def checkStationarity(data):
    pvalue = adfuller(data)[1]
    
    if(pvalue>0.05):
        msg = 'p-value={}. Data is not stationary'.format(pvalue)
    else:
        msg='p-value={}. Data is stationary'.format(pvalue)
        
    return(msg)
```

```python
ret=checkStationarity(train_df)
ret
```

```python
ret=checkStationarity(test_df)
ret
```
